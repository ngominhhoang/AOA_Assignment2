```python
def trace_route(n, k, c, f):
    order = []
    i = n
    while i != 0:
        minn = 1e15
        saved = -1
        for j in range(max(0, i-k), i):
            if f[j] + c[j] < minn:
                minn = f[j] + c[j]
                saved = j
        i = saved
        order.append(saved)
    order.reverse()
    return order
```

```python
# This file is only used to generate the plot. 
# It does not have to run on thunder (since it needs 3rd-party libraries)
# reference https://matplotlib.org/stable/gallery/lines_bars_and_markers/barchart.html

import random
from pathlib import Path
import time
import matplotlib.pyplot as plt
import numpy as np

from utils import read_input_basic

from task2a import task2a
from task2b import task2b
from task3 import task3
from task4 import task4
from task1 import alg1


algorithms = {
    #"strat0": alg1,
    #"strat1": task2a,
    #"strat2": task2b,
    "strat3": task3,
    "strat4": task4,
}

input_names = [Path(f).stem for f in Path("data/p1").glob("*.txt")]
results = {}
results_time = {}
for strat_name, f in algorithms.items():
    result = []
    result_time = []
    for input_name in input_names:
        if int(input_name) < 20:
            continue
        #print(input_name)
        with open(f"data/p1/{input_name}.txt", "r") as f_in:
            start_time = time.time()
            n, k, c = read_input_basic(f_in)
            #print(strat_name)
            
            res, _ = f(n, k, c)
            if strat_name == 'strat0':
                res = res[n]
            
            result_time.append(time.time() - start_time)
            result.append(res)
        
    results[strat_name] = result
    results_time[strat_name] = result_time
```

```python
# importing package 
import matplotlib.pyplot as plt 
import numpy as np 
  
# create data
n_values = list(range(10, 30, 2))
plt.plot(n_values, results_time['strat0'], label = "Algorithm 1")
plt.plot(n_values, results_time['strat1'], label = "Algorithm 2a", color = 'r') 
plt.plot(n_values, results_time['strat2'], label = "Algorithm 2b", color = 'b') 
plt.plot(n_values, results_time['strat3'], label = "Algorithm 3", color = 'orange') 
plt.plot(n_values, results_time['strat4'], label = "Algorithm 4", color = 'g') 
plt.xlabel('The number of platforms', fontsize=16)
plt.ylabel('The running time', fontsize=16) 
plt.rcParams.update({'font.size': 12})
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
 
plt.legend()
plt.savefig("plot/Task1_small.pdf", dpi = 300)
plt.show()
```
